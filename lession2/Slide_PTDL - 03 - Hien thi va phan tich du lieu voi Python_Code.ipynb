### Khởi tạo NumPy Array (Trang 16-18, 22-25)

```python
import numpy as np

# Tạo array từ list
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = a * b # Phép nhân element-wise
print(c)
```

```python
import numpy as np

a = np.zeros((2,2))   # Create an array of all zeros
print("Zeros array:\n", a)

b = np.ones((1,2))    # Create an array of all ones
print("\nOnes array:\n", b)

c = np.full((2,2), 7)  # Create a constant array
print("\nFull array:\n", c)

d = np.eye(2)         # Create a 2x2 identity matrix
print("\nIdentity matrix:\n", d)

e = np.random.random((2,2))  # Create an array filled with random values
print("\nRandom array:\n", e)
```

```python
import numpy as np

a = np.arange(15).reshape(3, 5)
print("Reshaped array:\n", a)
print("\nShape:", a.shape)
print("Dimensions:", a.ndim)
print("Data type:", a.dtype.name)
```

```python
# Tạo mảng 2 chiều
a = np.array([[0, 1, 2, 3], 
              [10, 11, 12, 13], 
              [20, 21, 22, 23]])
print("2D Array:\n", a)
print("\nDimensions:", a.ndim)
print("Shape:", a.shape)
```

### Ví dụ ứng dụng trong Tài chính (Trang 45)

```python
import pandas as pd

# Create a DataFrame to store transactions
columns = ['Date', 'Description', 'Amount']
ledger = pd.DataFrame(columns=columns)

# Function to add a transaction
def add_transaction(date, description, amount):
    global ledger
    new_transaction = pd.DataFrame([[date, description, amount]], columns=columns)
    ledger = pd.concat([ledger, new_transaction], ignore_index=True)

# Function to view the ledger
def view_ledger():
    print("--- Ledger ---")
    print(ledger)

# Function to get the current balance
def get_balance():
    return ledger['Amount'].sum()

# Adding sample transactions
add_transaction('2025-11-01', 'Income', 1000)
add_transaction('2025-11-02', 'Groceries', -200)
add_transaction('2025-11-03', 'Utilities', -100)

# Viewing the ledger
view_ledger()

# Checking the current balance
print("\nCurrent Balance:", get_balance())
```

## 3. Phân tích & Trực quan hóa Dữ liệu Iris

### Tải và khám phá dữ liệu (Trang 69-74)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

%matplotlib inline
```

```python
# Tải dữ liệu từ URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
df = pd.read_csv(url, names=names)

# Xem 10 dòng đầu tiên
print("--- First 10 rows ---")
print(df.head(10))
```

```python
# Xem 10 dòng cuối cùng
print("--- Last 10 rows ---")
print(df.tail(10))
```

```python
# Thống kê mô tả
print("--- Descriptive Statistics ---")
print(df.describe())
```

```python
# Thông tin tổng quan và kích thước
print("--- DataFrame Info ---")
df.info()
print("\n--- DataFrame Shape ---")
print(df.shape)
```

```python
# Phân phối của các lớp
print("--- Class Distribution ---")
print(df.groupby('class').size())
```

### Trực quan hóa dữ liệu (Trang 75-78)

```python
# Biểu đồ hộp (Box plot)
plt.rcParams["figure.figsize"] = (10,8)
df.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.suptitle('Box Plot for each feature')
plt.show()
```

```python
# Biểu đồ tần suất (Histogram)
df.hist()
plt.suptitle('Histogram for each feature')
plt.show()
```

```python
# Ma trận biểu đồ phân tán (Scatter plot matrix)
scatter_matrix(df)
plt.suptitle('Scatter Plot Matrix')
plt.show()
```

```python
# Pair plot với Seaborn
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, hue="class").fig.suptitle('Seaborn Pair Plot', y=1.02)
```
